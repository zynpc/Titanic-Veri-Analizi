```python
import pandas as pd
df = pd.read_csv('/kaggle/input/titanic/train.csv')
df.head()
```

1.Kazada ölenlerin yaş ortalamasını bulunuz

```python
avg_age_dead = df[df['Survived'] == 0]['Age'].mean()
print(f"Kazada ölenlerin yaş ortalaması: {avg_age_dead:.2f}")
```

 2. Kazada ölenlerin bilet fiyatlarının ortalaması ve medyanı

```python
dead_fare_mean = df[df['Survived'] == 0]['Fare'].mean()
dead_fare_median = df[df['Survived'] == 0]['Fare'].median()
print(f"Kazada ölenlerin bilet fiyatı ortalaması: {dead_fare_mean:.2f}")
print(f"Kazada ölenlerin bilet fiyatı medyanı: {dead_fare_median:.2f}")
```

3. Kazada ölen erkeklerin yaş ortalaması

```python
dead_male_age_mean = df[(df['Survived'] == 0) & (df['Sex'] == 'male')]['Age'].mean()
print(f"Kazada ölen erkeklerin yaş ortalaması: {dead_male_age_mean:.2f}")
```

4. Kazada ölen kadınların yaş ortalaması

```python
dead_female_age_mean = df[(df['Survived'] == 0) & (df['Sex'] == 'female')]['Age'].mean()
print(f"Kazada ölen kadınların yaş ortalaması: {dead_female_age_mean:.2f}")
```

5. Kazadan kurtulanların yaş ortalaması

```python
survived_age_mean = df[df['Survived'] == 1]['Age'].mean()
print(f"Kazadan kurtulanların yaş ortalaması: {survived_age_mean:.2f}")
```

6. Kazadan kurtulanların bilet fiyatlarının ortalaması

```python
survived_fare_mean = df[df['Survived'] == 1]['Fare'].mean()
print(f"Kazadan kurtulanların bilet fiyatı ortalaması: {survived_fare_mean:.2f}")
```

 7. Kazadan kurtulan toplam kişi sayısı

```python
total_survivors = len(df[df['Survived'] == 1])
print(f"Kazadan kurtulan toplam kişi sayısı: {total_survivors}")
```

8. 10 yaşından küçüklerin bilet fiyatlarının medyanı

```python
median_fare_under10 = df[df['Age'] < 10]['Fare'].median()
print(f"10 yaşından küçüklerin bilet fiyatı medyanı: {median_fare_under10:.2f}")
```

9. 1., 2., 3. sınıf bilet fiyatlarının ortalama ve medyan karşılaştırması

```python
fare_stats = df.groupby('Pclass')['Fare'].agg(['mean', 'median'])
print("Bilet sınıfına göre ortalama ve medyan bilet fiyatları:")
print(fare_stats)
```

 10. Kazada ölen kadın ve erkek oranlarını karşılaştırma

```python
total_men = len(df[df['Sex'] == 'male'])
dead_men = len(df[(df['Sex'] == 'male') & (df['Survived'] == 0)])
dead_men_ratio = dead_men / total_men

total_women = len(df[df['Sex'] == 'female'])
dead_women = len(df[(df['Sex'] == 'female') & (df['Survived'] == 0)])
dead_women_ratio = dead_women / total_women

print(f"Kazada ölen erkeklerin oranı: {dead_men_ratio:.2%}")
print(f"Kazada ölen kadınların oranı: {dead_women_ratio:.2%}")
```
